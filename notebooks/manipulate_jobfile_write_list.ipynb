```python
import numpy as np
import yaml
import os
from typing import Callable
```

```python
list_file = "/mnt/c/Users/Public/Documents/Uni/master/masterarbeit/ampel/Ampel-HU-astro/scripts/jobfileWriter/calibrateKilonovaEval_events.yml"

content = yaml.safe_load(open(list_file))

content
```

```python
def ManipulateContent(key: str, func: Callable, input: dict) -> dict:
    changed = input

    for entry, entry_dict in input.items():
        entry_dict[key] = func(entry_dict[key])

    return changed


def YearOffset(input_date: float):
    return input_date + 365
```

```python
year_off_output = ManipulateContent("trigger_jd", YearOffset, content)

new_output_dir = "/mnt/c/Users/Public/Documents/Uni/master/masterarbeit/ampel/Ampel-HU-astro/examples/O3_offset_jobfiles"
new_output_file = os.path.join(new_output_dir, "O3_offset_events.yml")

if not os.path.exists(new_output_dir):
    os.makedirs(new_output_dir)

with open(new_output_file, "w") as file:
    yaml.dump(year_off_output, file)

file.close()
```
